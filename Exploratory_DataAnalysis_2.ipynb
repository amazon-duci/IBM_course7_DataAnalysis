<h1>Exploratory Data Analysis - Laptops Pricing dataset</h1>

```python
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

%matplotlib inline
```

```python
def download(url, filename):
    response = requests.get(url)
    with open(filename, "wb") as f:
        f.write(response.content)
```

```python
filepath="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod2.csv"
```

```python
download(filepath, "laptops.csv")
file_name="laptops.csv"
```

```python
df = pd.read_csv(file_name, header=0)
df.head()
```

<h3>Task 3 - GroupBy and Pivot Tables</h3>
<p>Group the parameters "GPU", "CPU_core" and "Price" to make a pivot table and visualize this connection using the pcolor plot.</p>

```python
# To create the group
df_gptest = df[['GPU','CPU_core','Price']]
grouped_test1 = df_gptest.groupby(['GPU','CPU_core'], as_index=False).mean()

# To create the pivot table
grouped_pivot = grouped_test1.pivot(index='GPU', columns='CPU_core', values='Price')

# Create the plot
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

# Label names
col_labels = grouped_pivot.columns
row_labels = grouped_pivot.index

# Move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

# Insert labels
ax.set_xticklabels(col_labels, minor=False)
ax.set_yticklabels(row_labels, minor=False)

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Add color bar
fig.colorbar(im)

plt.show()
```

<h3>Task 4 - Pearson Correlation and p-values</h3>
<p>Use the scipy.stats.pearsonr() function to evaluate the Pearson Coefficient and the p-values for each parameter tested above.<br>This will help you determine the parameters most likely to have a strong effect on the price of the laptops.</p>

```python
for param in ['RAM_GB','CPU_frequency','Storage_GB_SSD','Screen_Size_inch','Weight_pounds','CPU_core','OS','GPU','Category']:
    pearson_coef, p_value = stats.pearsonr(df[param], df['Price'])
    print(param)
    print("The Pearson Correlation Coefficient for ",param," is", pearson_coef, " with a P-value of P =", p_value)
```
