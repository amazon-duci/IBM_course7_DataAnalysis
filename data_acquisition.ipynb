```python
file_path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv'
```

<p>To obtain the dataset, utilize the download() function as defined above:</p>

```python
# await download(file_path, "auto.csv")

# Because I'm using requests, the "await" is invalid
download(file_path, "auto.csv")
file_name="auto.csv"
```

```python
df = pd.read_csv(file_name)
```

```python
# show the first 5 rows using dataframe.head() method
print("The first 5 rows of the dataframe") 
df.head(5)
```

### Adding Headers
<p>First, create a list "headers" that include all column names in order. Then, use dataframe.columns = headers to replace the headers with the list you created.</p>

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#Replace headers and recheck our data frame:
df.columns = headers
df.columns
```

```python
df.head(10)
```

<p>Now, we need to replace the "?" symbol with NaN so the dropna() can remove the missing values:</p>

```python
df=df.replace('?',np.NaN)
```

```python
df.head(5)
```

<p>You can drop missing values along the column "price" as follows:</p>

<p><b>dropna():</b> means “drop” or “remove” any row (line) that has a missing value. <br>

<b>subset=["price"]:</b> tells it only look at the "price" column — we don’t care if other columns are missing. <br>

<b>axis=0:</b> means apply this rule to rows (not columns).</p>

```python
df=df.dropna(subset=["price"], axis=0)
df.head(20)
```

## Saving Dataset
<p>where index = False means the row names will not be written.</p>

```python
df.to_csv("automobile.csv", index=False)
```
