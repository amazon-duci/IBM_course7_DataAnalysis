<h1>Importing Dataset - Laptops Pricing</h1>

```python
import pandas as pd
import numpy as np
```

```python
import requests

#refer to data_acquisition file for explanation
def download(url, filename):
    response = requests.get(url)
    with open(filename, 'wb') as f:
        f.write(response.content)
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv"
```

```python
download(file_path, "laptops.csv")
file_name="laptops.csv"
```

```python
df = pd.read_csv(file_name, header = None) # removes the header but header = 0, loads using first row as header
df.head()
```

```python
#Adding headers
headers = [
    "Manufacturer", "Category", "Screen", "GPU", "OS", "CPU_core", "Screen_Size_inch",
    "CPU_frequency", "RAM_GB", "Storage_GB_SSD", "Weight_kg", "Price"
]
```

```python
df.columns = headers
df.head()
```

```python
# Replacing NaN values
df.replace('?',np.nan, inplace = True)
```
